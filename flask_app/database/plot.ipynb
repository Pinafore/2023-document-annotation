```python
def extrac_response_time(user_id, db_name):    
    conn = sqlite3.connect(db_name)
    cursor = conn.cursor()

    # Specify the user_id for which you want to retrieve the data
    # user_id = 43   # Replace with the actual user_id you want to retrieve data for

    # Execute the SQL query to retrieve the data
    cursor.execute('''SELECT id, label, doc_id, response_time, actual_label, local_training_acc, purity, 
                    local_testing_acc, RI, NMI, test_purity, test_RI, test_NMI
                    FROM recommendations
                    WHERE user_id = ?''', (user_id,))

    # Fetch all the rows
    results = cursor.fetchall()

    # Close the cursor and connection
    cursor.close()
    conn.close()
    result = []
    # Print or process the retrieved data as needed
    for index, row in enumerate(results, start=1):
        if index == 10:
            print(f"Row {index}: {row}")
        result.append(row)
    
    return result
```

```python
def time_to_seconds(time_str):
    hours, minutes, seconds = map(int, time_str.split(':'))
    total_seconds = hours * 3600 + minutes * 60 + seconds
    return total_seconds
```

```python
def extract_information(lsts):
    m1, m2, m3, m4, m5, m6 = [],[],[],[],[],[]
    medium_time, mean_time = [], []
    for user_condition in lsts:
        times = [time_to_seconds(ele[3]) for ele in user_condition]
        # print(times)
        p1, r1, n1, p2, r2, n2 = [],[],[],[],[],[]
        # metrics = [(ele[6], ele[8], ele[9], ele[10], ele[11], ele[12]) for ele in user_condition]
        
        for ele in user_condition:
            if ele[6] > 0:
                p1.append(ele[6])
                r1.append(ele[8])
                n1.append(ele[9])
                p2.append(ele[10])
                r2.append(ele[11])
                n2.append(ele[12])

        medium_time.append(np.median(times))
        mean_time.append(np.mean(times))
        m1.append(p1)
        m2.append(r1)
        m3.append(n1)
        m4.append(p2)
        m5.append(r2)
        m6.append(n2)

    return medium_time, mean_time, m1, m2, m3, m4, m5, m6
```

```python
user_lsts = [user6_active, user4_LDA, user1_SLDA, user6_SLDA, user6_ETM, user4_active]
med_time, mea_time, p1, r1, n1, p2, r2, n2 = extract_information(user_lsts)
```
